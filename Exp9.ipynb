```python
#importing required library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data =sns.load_dataset('titanic')
data
```

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

Here, we can see there are Null values in the dataset. Hence, we need to replace these values by mean (in case of numerical variables) or mode (in case of categorical variables)

```python
data['age'] = data['age'].fillna(np.mean(data['age']))
data['deck'] = data['deck'].fillna(data['deck'].mode()[0])
data['embark_town'] = data['embark_town'].fillna(data['embark_town'].mode()[0])
data['embarked'] = data['embarked'].fillna(data['embarked'].mode()[0])
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(10, 6))
palette = {0: "red", 1: "green"}
sns.boxplot(data=data, x='sex', y='age', hue='survived',palette = palette)
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper right')
plt.show()
```
