```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =sns.load_dataset('titanic')
data
```

```python
data.shape
```

```python
data.describe(include = 'object')
```

```python
sns.distplot(x = data['age'], bins = 10)
```

```python
sns.distplot(data['age'], bins = 10,kde=False)
```

```python
sns.jointplot(x = data['age'], y = data['fare'], kind ='scatter')
```

```python
sns.jointplot(x = data['age'], y = data['fare'], kind = 'hex')
```

```python
sns.barplot(x='sex', y='age', data=data)
```

```python
sns.boxplot(data['pclass'])
```

```python
sns.boxplot(data['sibsp'])
```

```python
sns.catplot(x= 'pclass', y = 'age', data=data, kind = 'box')
```

```python
sns.catplot(x= 'pclass', y = 'fare', data=data, kind = 'strip')
```

```python
sns.catplot(x= 'sex', y = 'fare', data=data, kind = 'strip')
```

```python
sns.catplot(x= 'sex', y = 'age', data=data, kind = 'strip')
```

```python
sns.pairplot(data)
```

```python
sns.scatterplot(x = 'fare', y = 'pclass', hue = 'survived', data = data)
```

```python
sns.catplot(x='pclass', y='fare', data=data, kind='bar')
```

```python
# Plotting the histogram of ticket prices
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='fare', bins=30, kde=True)
plt.title('Distribution of Ticket Prices on Titanic')
plt.xlabel('Ticket Fare')
plt.ylabel('Frequency')
plt.show()
```
